```python
import pandas as pd
import json
```

```python
county_df = pd.read_csv('./county_fips.csv')
```

```python
county_to_fips = {}
for i, r in county_df.iterrows():
    # county_to_fips[r['name']] = f"{(r['fips'] % 1000):003}"
    county_to_fips[r['county_name'] + "_" + r['state_code']] = f"{r['id'] % 84000000:00005}"
```

```python
county_to_fips
```

```python
with open("county_to_fips.json", "w") as outfile: 
    json.dump(county_to_fips, outfile)
```
